Обязательная часть
Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм со столбцами: <дата> - <заголовок> - <ссылка>.

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import time
time.sleep(0.2)
```

```python
import pandas as pd
```

```python
keywords = ['python', 'парсинг', 'Ignite']
```

```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
time.sleep(0.3)
```

```python
posts = soup.find_all('article', class_='post post_preview')
time.sleep(0.2)
posts
```

```python
import datetime
```

```python
from datetime import timedelta
```

```python
keywords_news = pd.DataFrame()
for post in posts:
    if any([keyword in post.find('div', class_='post__text').text for keyword in keywords]):
        date = post.find('span', class_='post__time').text
        date_element = date.split(' ')
        if 'сегодня'in date_element:
            date = datetime.datetime.today().date()
        elif 'вчера' in date_element:
            date = datetime.datetime.today().date()-timedelta(days=1)
        title = post.find('h2', class_='post__title').text
        link = post.find('a', class_='post__title_link').attrs.get('href')
        row = {'date': date, 'title': title, 'link': link}
        keywords_news = pd.concat([keywords_news, pd.DataFrame([row])])  
keywords_news
```
